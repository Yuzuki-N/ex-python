```python
src = 'https://wap.zol.com.cn/top/cell_phone/hot.html'
```

```python
import requests
```

```python
req = requests.get(src)
```

```python
req.text
```

```python
import re
```

```python
result = re.findall('<p class="pro-info-name f28">(.*?)<\/p>[\S\s]*?<span class="pro-info-price f24">(.*?)<\/span>',
          req.text, )
```

```python
with open('mobile.txt', 'w') as f:
    for item in result:
        f.write(item[0] + ' ' +  item[1] + '\n')
```
